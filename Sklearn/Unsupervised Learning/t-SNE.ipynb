# t-SNE for 2-dimensional maps
- t-SNE = “t-distributed stochastic neighbor embedding”
- Maps samples to 2D space (or 3D)
- Map approximately preserves nearness of samples
- Great for inspecting datasets
- t-SNE features are different every time
- However the varieties (=colors) have same position relative to one another

```python
df = pd.read_csv(r"datasets/grains.csv", header=None)
samples = df.iloc[:, :-1].values
variety_numbers = df.iloc[:, -1].values
```

```python
# Import TSNE
from sklearn.manifold import TSNE

# Create a TSNE instance: model
# try values between 50 and 200
# Wrong choice can lead to have coordinates bunch together
model = TSNE(learning_rate=200)

# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(samples)

# Select the 0th feature: xs
xs = tsne_features[:, 0]

# Select the 1st feature: ys
ys = tsne_features[:, 1]

# Scatter plot, coloring by variety_numbers
colors = {1: "black", 2: "red", 3: "pink"}
for idx in range(xs.shape[0]):
    plt.scatter(xs[idx], ys[idx], c=colors[variety_numbers[idx]], alpha=0.8)
plt.show()
```
